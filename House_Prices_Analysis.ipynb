# Exploratory Data Analysis of Ames Housing Data

The Ames Housing dataset is a comprehensive dataset that contains information on various properties in Ames, Iowa. It consists of 79 explanatory variables describing aspects of residential homes, such as size, type, neighborhood, and other property features. The target variable for this analysis is `SalePrice`, which represents the sale price of each house.

In this exploratory data analysis, we aim to:
- Understand the structure of the dataset and the relationships between features.
- Examine the distribution of the target variable, `SalePrice`.
- Handle missing values and outliers.
- Analyze correlations between numerical features and `SalePrice`.
- Visualize key insights to guide further model development or decision-making.
  
EDA is an approach to data analysis that uses a variety of techniques, largely visual, to: Maximize insight, reveal underlying structure, check for outliers, test assumptions, and determine optimal factors. It is an approach to analysis of your data that delays the usual assumptions and instead allows the data to speak for itself.

```python
#import packages
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Counting missing values

```python
missing = data.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
missing.plot.bar()
```

```python
data['SalePrice'].describe()
```

## Distribution Curve

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.histplot(data['SalePrice'],kde=True,bins=20);
```

```python
data['SalePrice'].describe()
```

## Correlation

```python
numeric_features = data.select_dtypes(include=[np.number])

numeric_features.columns
```

```python
categorical_features = data.select_dtypes(include=[object])

categorical_features.columns
```

```python
correlation = numeric_features.corr()
print(correlation['SalePrice'].sort_values(ascending=False),'\n')
```

```python
f , ax = plt.subplots(figsize = (14,12))

plt.title('Correlation of Numeric Features with Sale Price', y=1, size=16)

sns.heatmap(correlation,square = True, vmax= 0.8)
```

```python
k =11
cols = correlation.nlargest(k, 'SalePrice')['SalePrice'].index
print(cols)
cm = np.corrcoef(data[cols].values.T)
f , ax = plt.subplots(figsize = (14,12))
sns.heatmap(cm, vmax=0.8, linewidths=0.01, square=True, annot=True,cmap='viridis',
            linecolor="white",xticklabels = cols.values ,annot_kws = {'size':12},yticklabels = cols.values)
```

```python
first_quartile = data['SalePrice'].quantile(0.25)
third_quartile = data['SalePrice'].quantile(0.75)
IQR = third_quartile - first_quartile
```
